```python
from mpl_toolkits.mplot3d import Axes3D

def fun(x, y):
    return -(x+5*y)/10

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-10, 10, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z, color='blue', alpha=0.2)

ax.set_xlim([-15,15])
ax.set_ylim([-15,15])
ax.set_zlim([-15,15])
ax.quiver(0,0,0,x_1[0],x_1[1],x_1[2])

# A vector on the plane:
v = np.array([-5,1,0])
ax.quiver(0,0,0,v[0],v[1],v[2])


ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')


# rotate the axes and update
#ax.view_init(25, -100)
```

#### Systems of linear equations

```python
var('x y z')

x = np.array([x,y,z])
x
```

```python
A = np.array([[1,0,2],
              [1,1,0],
              [1,2,3]])

A@x
```

```python
b = np.array([1,2,3])
```

```python
np.linalg.solve(A,b)
```

```python
np.linalg.inv(A)@b
```

```python
x = np.arange(-5,5,0.05)
y = np.arange(-5,5,0.05)
z = np.arange(-5,5,0.05)

Y,Z = np.meshgrid(y,z)
X1 = 1-2*Z
X2 = 2-Y
X3 = 3-2*Y-3*Z

# plot the surface
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')


ax.plot_surface(X1, Y, Z,  color='blue',alpha=0.2)
ax.plot_surface(X2, Y, Z,  color='purple',alpha=0.2)
ax.plot_surface(X3, Y, Z,  color='green',alpha=0.2)

ax.scatter3D(1,1,0, color = "red", label='Solution')
plt.legend()

plt.show()
```
