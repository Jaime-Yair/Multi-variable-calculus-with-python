# Plotting a function of two variables

```python
plt.figure(figsize=(6,6))

ax = plt.axes(projection="3d")

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(x,y)
Z = 1-X**2 -Y**2

ax.plot_surface(X,Y,Z, alpha = 0.4)
plt.show()
```

# Plotting a contow plot

```python
x = np.linspace(-5,5,1000)
y = np.linspace(-5,5,1000)

X, Y = np.meshgrid(x,y)
Z = Y**2 - X**2

fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
plt.figure(figsize=(6,6))

fig.colorbar(cp) 
plt.show()
```

# Derivatives

```python
from scipy.misc import derivative

def f(x):
    return x**2

def deriva(x):
    return derivative(f,x)

x = np.linspace(-5,5,1000)

fig = plt.figure(figsize=(5,5))
plt.plot(x, f(x), label='function')
plt.plot(x,deriva(x),label='derivative')
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')

plt.legend()
plt.show()
```

### The derivate at a specific point

```python
# Define function
def f(x):
    return x**2

# Define the derivative
def deriva(x):
    return derivative(f,x)

# Define x data range 
x = np.linspace(-5,5,50)

# Choose point to plot tangent line
t1 = 3
y1 = f(t1)

m  = deriva(t1)

print("cordinates of the tangent: ",(t1,y1)) 
print("The slope at the chosen point: ", m)

# Define tangent line
# y = m*(x - x1) + y1
def line(x, t1, y1):
    return m*(x - t1) + y1

# Define x data range for tangent line
xrange = np.linspace(t1-1, t1+1, 10)

# Plot the figure
plt.figure(figsize=(7,5))
plt.plot(x, f(x))
plt.scatter(t1, y1, color='C1', s=50)
plt.plot(xrange, line(xrange, t1, y1), 'C1--', linewidth = 2)
plt.axvline(x=0, alpha=0.2, linestyle='--')
plt.axhline(y=0, alpha=0.2, linestyle='--')
```
