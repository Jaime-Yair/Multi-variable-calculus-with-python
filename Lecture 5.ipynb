```python
x = np.arange(-5,5,0.05)
y = np.arange(-5,5,0.05)
z = np.arange(-5,5,0.05)

Y,Z = np.meshgrid(y,z)
X = 7-2*Y+4*Z

# plot the surface
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X, Y, Z,  color='blue',alpha=0.2)

Q0 = [-1,2,2]
Q1 = [1,3,-1]

ax.scatter3D(Q0[0],Q0[1],Q0[2], color = "blue", alpha=0.4, label = "Point above the plane")
ax.scatter3D(Q1[0],Q1[1],Q1[2], color = "red", alpha=0.4, label = "Point below the plane")
ax.scatter3D(0,5/2,1/2, color='black', alpha=0.8, label='Point on the plane')
plt.legend()


t = np.linspace(-1,3,100)
f = []
for i in t:
    f.append(np.array([-1,2,2])+i*np.array([2,1,-3]))

for i in f:
    ax.scatter3D(i[0],i[1],i[2], color='blue', alpha=0.1, marker= '.',linewidth=0.3)


plt.show()
```
